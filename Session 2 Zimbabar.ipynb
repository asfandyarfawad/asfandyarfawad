```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
babar = pd.read_csv("babar.csv")
babar.head()
```

```python
# Setting Axis

x = babar["Opposition"]
y = babar["Runs"]

# Setting Fig Size

fig , ax = plt.subplots(figsize = (10,5)) # width , length

# Horizontal Bar Plot

ax.barh(x, y)

# Removing Spines/Border

# ax.spines["top"].set_visible(False)
# ax.spines["bottom"].set_visible(False)

pos = ["right" , 'left' , 'top' , 'bottom']
for i in pos:
    ax.spines[i].set_visible(False)

# Removing Ticks (Removing x , y ticks)
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')

# Adding Grid

ax.grid(visible = True , color = 'grey' , alpha = 0.2)

# Sorting Ascending order (Lexicographically)
ax.invert_yaxis()

# Add plot title

ax.set_title("Babar Azam Total Runs" , loc ='left' , fontdict = {'fontsize' : 20})

# Displaying Graph

plt.show()
```
