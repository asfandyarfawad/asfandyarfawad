```python
import pandas as pd
```

```python
df = pd.read_excel('Individual Details July-2023.xlsx')
```

```python
df
```

```python
df = df.drop(df.iloc[:,range(6,14)],axis=1)
```

```python
df.loc[:,'Date'] = pd.to_datetime(pd.to_datetime(df.loc[:,'Date']).dt.strftime('%Y-%m-%d'))
```

```python
pattern = '%H:%M'
df['Time In'] = pd.to_datetime(df['Time In'],format=pattern)
```

```python
late = df.loc[(df['Time In'] >  '1900-01-01 09:45:00') & (df['Time In'] < '1900-01-01 10:30:00')]
```

```python
half_days = df.loc[(df['Time In'] >  '1900-01-01 10:30:00')]
```

```python
late_1 = late.loc[:,['Date','Name','Time In']]
```

```python
late_1['Time In'] = late_1['Time In'].dt.strftime('%H:%M')
```

```python
late_1
```

```python
half_days_1 = half_days.loc[:,['Date','Name','Time In']]
half_days_1['Time In'] = half_days_1['Time In'].dt.strftime('%H:%M')
```

```python
half_days_1
```

```python
late_1.to_excel('Lates.xlsx', index=None)
```

```python
half_days_1.to_excel('Half Days.xlsx', index=None)
```
